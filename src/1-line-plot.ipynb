### Exercise 1:line plot


```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.arange(0, 11)  # Creates an array [0, 1, 2, ..., 10]

# Compute y values (square of x)
y = x ** 2

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')

# Add labels and title
plt.xlabel("X Values")
plt.ylabel("Y Values (x²)")
plt.title("Line Plot of y = x²")

# Add grid for better readability
plt.grid(True, linestyle="--", alpha=0.6)

# Show legend
plt.legend()

# Display the plot
plt.show()
```
